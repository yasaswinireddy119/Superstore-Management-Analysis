```python
clean_df["Order Date"] = pd.to_datetime(
    clean_df["Order Date"],
    format="%d-%m-%Y"
)

clean_df["Ship Date"] = pd.to_datetime(
    clean_df["Ship Date"],
    format="%d-%m-%Y"
)
```

```python
clean_df.columns = (
    clean_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("%", "percent")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)
```

```python
clean_df["auto_reorder"] = (
    clean_df["auto_reorder"]
    .str.strip()
    .str.upper()
)
```

```python
clean_df["shipping_days"] = (
    clean_df["ship_date"] - clean_df["order_date"]
).dt.days
```

```python
clean_df["order_year"] = clean_df["order_date"].dt.year
```

```python
clean_df["order_month_name"] = (
    clean_df["order_date"].dt.month_name()
)
```

```python
clean_df["order_day"] = (
    clean_df["order_date"].dt.day_name()
)
```

```python
clean_df["profit_margin"] = (
    clean_df["profit"] /
    clean_df["sales_amount"]
) * 100
```

```python
clean_df["revenue_per_unit"] = (
    clean_df["sales_amount"] /
    clean_df["quantity"]
)
```

```python
clean_df["inventory_status"] = np.where(
    clean_df["stock_left"] <= 20,
    "Low Stock",
    "Sufficient Stock"
)
```

## Business Question 1

### Which Customer Segment places the most orders?

**Objective:**

Analyze the distribution of orders across different customer segments to identify the company's primary customer base.

### Business Insight

- The customer segment with the highest number of orders represents the company's primary customer base.
- The segment with fewer orders may require targeted marketing strategies to improve engagement.

### Recommendation

- Strengthen customer loyalty programs for the leading segment.
- Design promotional campaigns for lower-performing segments to increase their order volume.

# Multivariate Analysis

## Objective

The objective of this section is to analyze the relationship among multiple business variables simultaneously. This helps identify hidden patterns and supports strategic business decisions regarding sales, profit, customers, products, and regions.

```python
pivot_sales = clean_df.pivot_table(
    values="sales_amount",
    index="region",
    columns="category",
    aggfunc="sum"
)

plt.figure(figsize=(10,6))

sns.heatmap(
    pivot_sales,
    annot=True,
    fmt=".0f",
    cmap="YlGnBu"
)

plt.title("Sales by Region and Category")
plt.xlabel("Category")
plt.ylabel("Region")

plt.show()
```

### Observation

The heatmap compares total sales across different regions and product categories.
### Business Insight

- Darker cells represent higher sales.
- It becomes easy to identify which categories perform best in each region.

### Recommendation

Increase inventory and marketing efforts for high-performing region-category combinations while improving strategies for weaker combinations.

```python
pivot_profit = clean_df.pivot_table(
    values="profit",
    index="region",
    columns="category",
    aggfunc="sum"
)

plt.figure(figsize=(10,6))

sns.heatmap(
    pivot_profit,
    annot=True,
    fmt=".0f",
    cmap="Greens"
)

plt.title("Profit by Region and Category")
plt.xlabel("Category")
plt.ylabel("Region")

plt.show()
```

### Observation

The heatmap compares profitability across different categories and regions.

### Business Insight

High sales do not necessarily indicate high profitability. Some categories may have strong revenue but lower profit margins.

### Recommendation

Focus on expanding region-category combinations that generate consistently high profits.

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=clean_df,
    x="region",
    y="sales_amount",
    hue="customer_segment",
    estimator="sum"
)

plt.title("Sales by Region and Customer Segment")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

# Module 8 – Business Insights & Recommendations

## Executive Summary

This project analyzed a Superstore Management dataset containing **1,000 customer orders** using Python, Pandas, NumPy, Matplotlib, and Seaborn. The dataset was cleaned, transformed, and enriched through feature engineering before performing exploratory data analysis to evaluate sales performance, profitability, customer behavior, product performance, regional trends, and operational efficiency.

The analysis revealed key business patterns that can support data-driven decision-making in sales strategy, inventory planning, customer relationship management, and business growth.

---

## Key Business Insights

### 1. Overall Business Performance
The company generated approximately **$12.33 million** in total sales and **$3.01 million** in total profit, indicating a healthy and profitable business with a profit margin of around **24%**.

### 2. Customer Behavior
The business processed **1,000 orders** from **595 unique customers**, showing that many customers placed multiple orders. This indicates a strong level of customer retention and repeat purchasing.

### 3. Regional Performance
Regional analysis showed noticeable differences in sales and profitability across locations. Certain regions consistently contributed more revenue and profit, while others demonstrated opportunities for business expansion and improved market penetration.

### 4. Product Category Performance
Product categories contributed differently to total revenue and profit. Some categories generated high sales volumes, while others delivered stronger profitability, emphasizing the importance of evaluating both sales and profit when making business decisions.

### 5. Customer Segment Analysis
Customer segments exhibited different purchasing patterns and revenue contributions. Identifying the highest-value customer segment enables the business to design personalized marketing campaigns and loyalty programs.

### 6. Sales and Profit Relationship
Higher sales did not always translate into higher profits. Factors such as discounts, pricing strategies, and operational costs influenced overall profitability, highlighting the need to monitor profit alongside revenue.

### 7. Payment Preferences
Customers used multiple payment methods, with some payment modes contributing a larger share of total sales. Understanding customer payment preferences helps improve the checkout experience and supports future payment optimization strategies.

### 8. Inventory and Delivery Operations
The analysis of stock availability, reorder indicators, shipping duration, and delivery status provided valuable insights into operational performance. Monitoring these metrics can help reduce stock shortages, improve inventory planning, and ensure timely deliveries.
